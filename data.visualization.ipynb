# Data Visualization 

```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# load the dataset

```python
# load the csv data into the Dataframe

df = pd.read_csv('C:/Users/Etimbuk/Downloads/user_data.csv')
df.head(11)
```

# Visualizing Ages with Highest Income

```python
# Group by age and calculate the average income
age_income = df.groupby('age')['total_income'].mean().sort_values(ascending=False) 
# Plot the data
plt.figure(figsize=(10, 6))
age_income.plot(kind='bar')
plt.title('Ages with the Highest Income')
plt.xlabel('Age')
plt.ylabel('Average Total Income')
plt.xticks(rotation=45)
```

```python
# Visualizing Ages with highest Income using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x="age", y="total_income", data=df, hue="age", palette="viridis", legend=False) 
plt.title('Ages with the Highest Income', fontsize=16)
# Save and Export Chart for Client Usage
plt.savefig('age_income_chart2.png')
plt.show() 
```

# Interpretation
-The visualizations shows that ages with higher incomes tend to be older and this is a realistic trend , as income levels increases with age due to factors such as career advancement and years of experience.


# Visualizing Gender distribution across spending categories

```python
# Calculate the total spending by category for each gender
gender_expenses = df.groupby('gender')[['expenses.utilities', 'expenses.entertainment', 'expenses.school_fees', 'expenses.shopping' , 'expenses.healthcare']].sum()

# Plot the data
gender_expenses.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Gender')
plt.ylabel('Total Spending')
plt.legend(title='Expense Categories')

# Save and Export Chart for Client Usage
plt.savefig('gender_expenses_chart.png')
```

# Interpretation
- These clustered chart shows that male gender has the highest expenditure on healthcare, while female  has the highest expenses on shool fee, and shopping which the primary expenditure surpassing other categories. Conversely, expenditures on entertainment appear notably lower for both genders
